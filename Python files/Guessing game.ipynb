### Guessing game

In this code, we shall be choosing a number at random and guessing the number

```python
import random
x = random.randrange(1,5)
guess = int(input("Enter any number: "))


while x != guess:
    if guess < x:
        print("guess is too low")
        guess = int(input("Enter a different number"))
    elif guess > x:
        print("guess is too high")
        guess = int(input("Enter a different number"))
    else:
        break

print("You guessed it right!")
```

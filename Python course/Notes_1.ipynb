## Boolean (bool)

These are special values that are neither a string nor a number, they are: `True`and `False`. They have no quotation marks around them and it is not a numeric value.

We can use `True` for situations like if a feature is on or if the data is available, eg

`powered_on = True`

We can store `True`in a variable just like a string or number. We can also display it:

```python
correct = True
print(correct)
```

We can negate the `False` and get `True` and vice versa. That is `not True = False`. Note also that we can negate the variable to get the negation of the the Boolean values.

```python
print(not correct)
```

We can save a whole negation in another variable too.

```python
morning = True
is_evening = not morning
print(is_evening)
```

To compare two entities, we use the equallity operator `==`. Wen comparing two things, we have two results; it is either `True`or `False`.

```python
entered_pin = 1334
expected_pin = 2113
print(entered_pin == expected_pin)
```

An inequality is given by `!=`.

We can store a comparison in a variable too, and print it out.

```python
result = 1!=2
print(result)
```

## Define (def) function

Now, we focus on functions themselves. Now, we want to be able to define our functions, using the keyword `def`

```python
def hello():
    print("hello")


name = input("What is your name?")
hello()

print(name)
```

Next, we customise our defined function.

```python
def hello(to):
    print("hello,", to)


name = input("What is your name?")
hello(name)
```

Now, in case we want a code that prints hello world, just in casee the programmer forgets to input values, we set it as follows

```python
def hello(to = "world"):
    print("hello,", to)


hello()
name = input("What is your name?")
hello(name)
```

In the above code, the defined code has been created. Imagine for a second that this code is defined much later, the code will not run properly. For example to see this, run the code bellow in python

`name = input("What is your name?")`

`hello(name)`


`def hello(to):`

`   print("hello,", to)`

How to fix it.

```python
def main():
    name = input("What is your name?")
    hello(name)


def hello(to = "world"):
   print("hello,", to)

main()
```
